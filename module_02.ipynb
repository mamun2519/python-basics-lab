#Functions

```python
# Positional argument
def calculateCostAndSelery(selery, cost):
  return int(selery) - int(cost)

result = calculateCostAndSelery("1000", 350)
print(result)
```

```python
# multi Positional argument ( you dont not how many argument sent)
def name(*names):
  print(names)
name(2 , 4 , 5)
```

```python
# multi keyword arugment(key value formet)
def studentInformation(**studentList):
  print(studentList)
  print(studentList.keys)
  for i in studentList:
    print(i)

studentInformation(name= "Mamun", age= "20", address= "DL")
```
