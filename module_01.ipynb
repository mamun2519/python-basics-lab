# Dictionaries

```python
userInfo = {
    "name": "mohammad Mamun",
    "age": 25,
    "gender": "Male",
    "phone": "01860700702",
    "marrite": False
}

## Access data
print(userInfo)
print(userInfo.get("age"))
print(userInfo.keys())

# change the iteam
userInfo['age'] = 27
userInfo.update({"marrite": True})

# remove iteam
userInfo.pop("phone")
userInfo.popitem()


print(userInfo)

## Access the value
for i in userInfo.values():
  print(i)
```
