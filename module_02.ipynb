#Functions

```python
# Positional argument
def calculateCostAndSelery(selery, cost):
  return int(selery) - int(cost)

result = calculateCostAndSelery("1000", 350)
print(result)
```

```python
# multi Positional argument ( you dont not how many argument sent)
def name(*names):
  print(names)
name(2 , 4 , 5)
```

```python
# multi keyword arugment(key value formet)
def studentInformation(**studentList):
  print(studentList)
  print(studentList.keys)
  for i in studentList:
    print(i)

studentInformation(name= "Mamun", age= "20", address= "DL")
```

# Iterator
it is object . its retune one by one value

```python
nums = [3, 4 , 5 , 6]
it = iter(nums)
print(next(it))
print(next(it))
```

# Generator
It is easy way to create iterator

```python
def my_gen():
  yield 1
  yield 2
  yield 3
g = my_gen()
print(next(g))
print(next(g))
```

# Lambda func
its is anonymous function

```python
totalNumber = lambda a,b : a +b
print(totalNumber(4,6))
```

# Map

```python
myList = [5, 6 , 9,10]
a = map(lambda a: a * 2, myList)
result = list(a)
print(result)
```

# filter


```python
nums = [1, 2, 3, 4, 5]

result = filter(lambda x: x % 2 == 0, nums)
print(list(result))  # [2, 4]
```
